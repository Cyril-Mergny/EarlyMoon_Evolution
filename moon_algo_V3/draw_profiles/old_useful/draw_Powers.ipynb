```python
path = "../Outputs/"
```

```python
%matplotlib inline  
#%matplotlib
import numpy as np  
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import data
filename = path+"Pcrust.txt"
data = pd.read_csv(filename, skipinitialspace=True, sep=" ",index_col=False,names=["P"])
Pcrust = np.array(data["P"])/1e12
```

```python
# Import data
filename = path+"Pflux_1pt.txt"
data = pd.read_csv(filename, skipinitialspace=True, sep=" ",index_col=False,names=["P"])
Pflux_1pt = np.array(data["P"])/1e12
```

```python
# Import data
filename = path+"Pflux_3pt.txt"
data = pd.read_csv(filename, skipinitialspace=True, sep=" ",index_col=False,names=["P"])
Pflux_3pt = np.array(data["P"])/1e12
```

```python
time = 1
Time =[time*k  for k in range(len(Pcrust))]
```

```python
fig, ax = plt.subplots(figsize=(7,8))

ax.plot(Time, Pcrust, label = "Pcrust")
ax.plot(Time, Pflux_1pt, label = "Ps (1point)")
ax.plot(Time, Pflux_3pt, label = "Ps (3points)")
ax.set_xlabel("Time (myears)")
ax.set_ylabel(r"Power (TW)")
plt.legend()


resolution = 100
plt.title("Timestep={0}my, resolution={1}".format(time,resolution))
plt.savefig("../figures/nrj_{0}my_{1}_Ps.png".format(time,resolution))
plt.show()
```
